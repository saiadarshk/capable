```python
df=pd.read_csv('C:\\Users\\pc\\python\\WEEK5\\Data.csv',header=None)
```

```python
eng=np.reshape(df[1].values,(13,1))
imputer= SimpleImputer(missing_values=np.nan,strategy='most_frequent')##mean,median,most_frequent
imputer.fit(eng)
df['eng']=imputer.transform(eng)

tel=np.reshape(df[2].values,(13,1))
imputer= SimpleImputer(missing_values=np.nan,strategy='most_frequent')##mean,median,most_frequent
imputer.fit(tel)
df['tel']=imputer.transform(tel)



hin=np.reshape(df[3].values,(13,1))
imputer= SimpleImputer(missing_values=np.nan,strategy='most_frequent')##mean,median,most_frequent
imputer.fit(hin)
df['hin']=imputer.transform(hin)

mat=np.reshape(df[4].values,(13,1))
imputer= SimpleImputer(missing_values=np.nan,strategy='most_frequent')##mean,median,most_frequent
imputer.fit(mat)
df['mat']=imputer.transform(mat)
```
