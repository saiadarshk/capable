```python
super_market={}
def create_new_super_market():
    print("*creating new super_market*")
    while True:
        product_id=input("Enter the product id(Enter 0 to exit):")
        if(product_id=="0"):
            break
        else:
            product_name=input("Enter the name of the product:  ")
            price= float(input("Enter product price(in ruppes): "))
            super_market[product_id] = {'name': product_name, 'price':price }
            print("super_market:", super_market)
            print("-"*100)
            print("*list of products*")
            print("~"*30)
            for i in   super_market.keys():
                print(i,":",super_market[i])
def show_details():
    print( "Showing Product Details")
    id=input("Enter the product id:")
    print(id,":",super_market[id])
def delet_details():
    print( "deleting  product from super_market")
    id=input("Enter the product id:")
    del super_market[id]
    print("*list of products*")
    print("~"*30)
    for i in   super_market.keys():
        print(i,":",super_market[i])
def main():
    while True:
        print("_"*50)
        print("1. Show Product Details")
        print("2. delete any product from super_market")
        print("3. create new super_market(super_market will restart!)")
        print("4. Exit")
        choice = int(input("Enter your choice: "))
        print("_"*50)
        if choice == 1:
            show_details()
        elif choice == 2:
            delet_details()
        elif choice == 3:
            create_new_super_market()    
        elif choice == 4:
            break
        else:
            print("Invalid choice! Please try again.")        
if __name__ == '__main__':
    main()
```
