```python
def typecast(col_name):
    
    lst=[]
    for i in df[col_name]:
        lst.append(int(i))
    df[col_name]=lst
    # here we are converting float values to int
```

```python
typecast('mat')
typecast('sci')
typecast('tel')
typecast('eng')
```
