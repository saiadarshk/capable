```python
df=pd.read_csv('C:\\Users\\Adarsh\\python\\WEEK5\\Data.csv',header =None)
name=df[1]
tel=df[2]
eng=df[3]
mat=df[4]
sci=df[5]

del df[0]
del df[1]
del df[2]
del df[3]
del df[4]
del df[5]

df['name']=name
df['tel']=tel
df['eng']=eng
df['mat']=mat
df['sci']=sci

print(df)
```
